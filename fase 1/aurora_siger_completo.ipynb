**Missão:** Ignition Zero — O Início da Aurora  
**Objetivo:** Validar parâmetros técnicos para autorizar ou abortar o lançamento da nave Aurora Siger.

## 1. Organização e Descrição da Telemetria

A telemetria é o conjunto de dados enviados automaticamente pelos sistemas embarcados da nave para monitoramento em tempo real. Durante a pré-decolagem, os seguintes parâmetros são monitorados:

| Parâmetro | Valor | Unidade | Faixa Segura | Significado |
|---|---|---|---|---|
| Temperatura interna | 22.5 | °C | 18°C – 30°C | Condição habitável e segura para os sistemas eletrônicos |
| Temperatura externa | -60.0 | °C | -120°C ≤ T ≤ 20°C | Condição esperada fora da nave; isolamento deve ser verificado |
| Integridade estrutural | 1 | binário (0/1) | 1 = OK | Confirma que a estrutura da nave está íntegra |
| Nível de energia | 87.3 | % | ≥ 80% | Carga atual das baterias; margem mínima para decolagem segura |
| Pressão do tanque | 350.0 | kPa | 300 – 400 kPa | Pressão dos propelentes; fora do range causa falha de ignição |
| Módulo de propulsão | 1 | binário (0/1) | 1 = ativo | Sistema responsável pelo impulso da nave |
| Módulo de comunicação | 1 | binário (0/1) | 1 = ativo | Sistema de transmissão de dados com a base terrestre |
| Módulo de navegação | 1 | binário (0/1) | 1 = ativo | Sistema de orientação e trajetória da nave |

> ⚠️ **Atenção:** O nível de energia (87.3%) está a apenas 7.3 pontos percentuais do limite mínimo de 80%. Monitoramento contínuo recomendado durante a contagem regressiva.

````python
# Dados de telemetria da nave Aurora Siger
telemetria = {
    "temperatura_interna": 22.5,   # °C — faixa segura: 18 a 30
    "temperatura_externa": -60.0,  # °C — faixa segura: -120°C ≤ T ≤ 20°C
    "integridade_estrutural": 1,   # 1 = OK, 0 = falha
    "nivel_energia": 87.3,         # % — mínimo: 80%
    "pressao_tanque": 350.0,       # kPa — faixa segura: 300 a 400
    "modulo_propulsao": 1,         # 1 = ativo, 0 = falha
    "modulo_comunicacao": 1,       # 1 = ativo, 0 = falha
    "modulo_navegacao": 1,         # 1 = ativo, 0 = falha
}

print("=" * 50)
print("  TELEMETRIA — NAVE AURORA SIGER")
print("=" * 50)
for chave, valor in telemetria.items():
    print(f"  {chave:<25} → {valor}")
print("=" * 50)
````

## 2. Algoritmo de Verificação

### 2.1 Pseudocódigo

```
INÍCIO verificar_decolagem(telemetria):
  falhas ← lista vazia

  SE temperatura_interna < 18 OU temperatura_interna > 30:
      adicionar "Temperatura interna fora do range" em falhas

  SE temperatura_externa < -120 OU temperatura_externa > 20:
      adicionar "Temperatura externa fora do range" em falhas

  SE integridade_estrutural ≠ 1:
      adicionar "Falha na integridade estrutural" em falhas

  SE nivel_energia < 80:
      adicionar "Energia abaixo do mínimo" em falhas

  SE pressao_tanque < 300 OU pressao_tanque > 400:
      adicionar "Pressão do tanque fora do range" em falhas

  PARA CADA modulo EM [propulsao, comunicacao, navegacao]:
      SE modulo ≠ 1:
          adicionar "Módulo com falha: " + modulo em falhas

  SE falhas está vazia:
      IMPRIMIR "✅ PRONTO PARA DECOLAR"
  SENÃO:
      IMPRIMIR "❌ DECOLAGEM ABORTADA"
      PARA CADA falha EM falhas:
          IMPRIMIR "→ " + falha

FIM
```
O algoritmo percorre 6 verificações em sequência:
### 2.2 Fluxo lógico
1. Temperatura interna (18°C ≤ T ≤ 30°C)
2. Temperatura externa (-120°C ≤ T ≤ 20°C)
3. Integridade estrutural (deve ser 1)
4. Nível de energia (deve ser ≥ 80%)
5. Pressão do tanque (300 ≤ P ≤ 400 kPa)
6. Status dos três módulos críticos (todos devem ser 1)

Se **todas** as condições forem satisfeitas → **PRONTO PARA DECOLAR**  

Se **qualquer** condição falhar → **DECOLAGEM ABORTADA** + lista de falhas  

---
## 3. Script em Python — Verificação Automatizada

````python
def verificar_decolagem(t):
    """
    Verifica os parâmetros de telemetria da nave Aurora Siger.
    Retorna 'PRONTO PARA DECOLAR' ou 'DECOLAGEM ABORTADA' com lista de falhas.

    Parâmetros verificados:
    - Temperatura interna: 18°C a 30°C
    - Temperatura externa: -120°C a 20°C
    - Integridade estrutural: deve ser 1 (OK)
    - Nível de energia: mínimo 80%
    - Pressão do tanque: 300 a 400 kPa
    - Módulos críticos (propulsão, comunicação, navegação): devem ser 1
    """
    falhas = []

    # Verificação 1: Temperatura interna
    if not (18 <= t["temperatura_interna"] <= 30):
        falhas.append(f"Temperatura interna fora do range (valor: {t['temperatura_interna']}°C)")

    # Verificação 2: Temperatura externa
    if not (-120 <= t["temperatura_externa"] <= 20):
        falhas.append(f"Temperatura externa fora do range (valor: {t['temperatura_externa']}°C)")

    # Verificação 3: Integridade estrutural
    if t["integridade_estrutural"] != 1:
        falhas.append("Falha na integridade estrutural")

    # Verificação 3: Nível de energia
    if t["nivel_energia"] < 80:
        falhas.append(f"Energia abaixo do mínimo (valor: {t['nivel_energia']}%)")

    # Verificação 4: Pressão do tanque
    if not (300 <= t["pressao_tanque"] <= 400):
        falhas.append(f"Pressão do tanque fora do range (valor: {t['pressao_tanque']} kPa)")

    # Verificação 5: Módulos críticos
    for modulo in ["modulo_propulsao", "modulo_comunicacao", "modulo_navegacao"]:
        if t[modulo] != 1:
            falhas.append(f"Módulo com falha: {modulo}")

    # Resultado final
    print("=" * 50)
    print("  RESULTADO DA VERIFICAÇÃO PRÉ-DECOLAGEM")
    print("=" * 50)

    if not falhas:
        print("  ✅ STATUS: PRONTO PARA DECOLAR")
        print("  Todos os parâmetros dentro das faixas seguras.")
    else:
        print("  ❌ STATUS: DECOLAGEM ABORTADA")
        print(f"  {len(falhas)} falha(s) detectada(s):")
        for f in falhas:
            print(f"    → {f}")

    print("=" * 50)

# Executar verificação
verificar_decolagem(telemetria)
````

### 3.1 Teste com cenário de falha

Para demonstrar o funcionamento do algoritmo em caso de parâmetros fora das faixas:

````python
# Cenário de teste: simulando falhas para validar o algoritmo
telemetria_com_falhas = {
    "temperatura_interna": 35.0,   # FORA DO RANGE (acima de 30°C)
    "temperatura_externa": -60.0,
    "integridade_estrutural": 1,
    "nivel_energia": 72.0,          # FORA DO RANGE (abaixo de 80%)
    "pressao_tanque": 350.0,
    "modulo_propulsao": 1,
    "modulo_comunicacao": 0,         # FALHA no módulo de comunicação
    "modulo_navegacao": 1,
}

verificar_decolagem(telemetria_com_falhas)
````

## 4. Análise Energética

````python
# Parâmetros energéticos da nave
capacidade_total_kwh = 500        # Capacidade total das baterias (kWh)
carga_atual_pct      = telemetria["nivel_energia"]  # 87.3%
consumo_decolagem_kwh = 120       # Consumo estimado durante a decolagem (kWh)
perda_estimada_pct   = 5          # Perdas por calor, conversão e distribuição (%)

# Cálculos
energia_disponivel  = capacidade_total_kwh * (carga_atual_pct / 100)
energia_apos_perda  = energia_disponivel * (1 - perda_estimada_pct / 100)
autonomia_restante  = energia_apos_perda - consumo_decolagem_kwh
autonomia_pct       = (autonomia_restante / capacidade_total_kwh) * 100

# Relatório energético
print("=" * 50)
print("  ANÁLISE ENERGÉTICA — AURORA SIGER")
print("=" * 50)
print(f"  Capacidade total:          {capacidade_total_kwh:.1f} kWh")
print(f"  Carga atual:               {carga_atual_pct:.1f}%")
print(f"  Energia disponível:        {energia_disponivel:.1f} kWh")
print(f"  Perdas estimadas ({perda_estimada_pct}%):    -{energia_disponivel - energia_apos_perda:.1f} kWh")
print(f"  Energia após perdas:       {energia_apos_perda:.1f} kWh")
print(f"  Consumo na decolagem:      -{consumo_decolagem_kwh:.1f} kWh")
print("-" * 50)
print(f"  Reserva pós-decolagem:     {autonomia_restante:.1f} kWh ({autonomia_pct:.1f}%)")
print("=" * 50)

# Avaliação da margem
if autonomia_restante > 200:
    print("  ✅ Margem energética adequada para a missão.")
elif autonomia_restante > 100:
    print("  ⚠️  Margem energética moderada — monitorar consumo.")
else:
    print("  ❌ Margem energética crítica — risco de falha de missão.")
````
